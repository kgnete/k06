From the two bank balance examples, we can see that the natural frequency of a first-order LDE, or equivalently , governs the evolution of its associated sequences. In particular,

    If $$ \alpha _1 > 1 $$ , the sequence grows monotonically and without bound, and we say that the first-order LDE is unstable.

    If $0 < \alpha _1 < 1$

    , the sequence decreases monotonically to zero, and we say that the first-order LDE is stable.

    If $\alpha _1 < 1$, then the sequence does not change value with index, and the first-order LDE is neither stable nor unstable.
