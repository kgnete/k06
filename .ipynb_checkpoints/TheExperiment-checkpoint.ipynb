As we noted in the above tutorial, the motorCmd variable in the Arduino sketch is linearly proportional to the fractional pulse width for the pulse-width modulated current in the motor, and is therefore linearly proportional to the average motor current. And for a typical DC motor (but not a brushless motor), motor current is proportional to the motor torque.

If there are no of friction forces, the motor angular acceleration is proportional to motor torque, so in the friction-free case, our motor's angular acceleration is proportional to the value of the Arduino variable motorCmd. But this contradicts what you should have already observed, that motorCmd must be reasonably large (in the range of 50 to 150) just to keep the motor spinning at a steady speed of 200 as measured by the Arduino.
