# Inputs and Commands

Holy goodness what is this new stuff? Well the top right still has the same old serial port things so you should know that already. There are more things to vary this time through...three sliders to be precise. They are:

    Desired Motor Speed: This is the value you'd like to see the motor running at; basically the input to your feedback system. You can set it at any value up to 255. There are notes in the main body of the lab about these units.
    Kp: This is a value that will be used for proportional speed control on the motor. This is the gain of the feedback loop.
    Direct: This is an "offset" value that added to the driving value determined by the feedback portion. It is basically the "open-loop" value you used in Week 1.

There is also a different looking slider with the label of Generate CSV?. This allows you the option to generate csv files of the data being plotted. When you Turn on the CSV generator it creates two csv files for you: one with named using Epoch Time for when you started and one named 'current.csv'. The former is an archival version and the latter is a most recent recording version (for your own convenience) that gets overwritten with every new recording (while leaving the archival one alone).

The CSV file is labeled for you with time stamp, the three values being plotted (see below) and the settings you are applying. Feel free to use this tool to analyze your system reponses in more detail (CSV files are totes compatible with Excel or any other generic spreadsheet program. They're also really easily readable in Python and other languages. We can discuss how to do this in the forum if you'd like :) ).

And as with all our stuff, keep in mind that while we tried to stress-test this new feature, it isn't super water-tight...like if you start writing a csv file and then leave the thing running for an hour you're going to come back with a csv file that is ~66,000 lines long or so...that may or may not make your system happy... so just keep this in mind while you're using it. The ultimate long-term goal with a feature like this is to allow us a means of autograding your system responses/behavior in version 2 of this course, so this is a part of our development for future labs. Any ideas/thoughts would be welcome as always in the discussion forums!

# Outputs/Plots

There are three plots generated in this new version of the software. They are:

    (Actual) Motor Speed: The measured value of the motor speed. The lab will go into more detail about how this is measured and what its units are.
    Error: This is difference (error) between the Desired Motor Speed (set by you in the slider above) and the Actual Motor Speed (what is maintained by the system)
    Motor Command: This is the signal being sent to the motor. In this lab it is the product of the error signal and the proportional gain .

For Lab2 the only real hardware change that needs to be done is ensuring that when driven by the control, the propeller forces itself downward. This is desired here since we want to focus on propeller speed only and not on speed when it is being affected by flying upwards out of control and whacking electronics. In moving from Lab 1, this should entail only the flipping of the placement of the two propeller wires! Once complete you will be able to run your prop on the edge of a table like shown below without fear of it flying away and messing up your measurements.
