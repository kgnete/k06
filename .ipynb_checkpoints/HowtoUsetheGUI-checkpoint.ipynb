# 6.302 Software
Assuming you've done everything you need to done all the tasks listed above and on previous pages, download the zip file provided, extract it, and then move into the newly extracted folder in your terminal that. Once there, if you're using Python 2, run the command (you may be prompted for a machine password based on your settings):

    sudo pip install -r req.txt
Or if you're a Python 3, do:

    sudo pip3 install -r req.txt
A whole bunch of text should fly by in a neutral color like green (if red stuff appears, it *may* mean there was an error in install. At that point I'd head to the forum and we'll try to see what we can figure out.).

Note: One bug I ran into when installing on my Ubuntu v15.04 machine was I didn't have developer packages installed...If you get an error, try running:

    sudo apt-get install python3-dev
    sudo apt-get install libevent-dev
Then try the pip install again!

If the install was a success, the next thing you need to do is start the server using Python. If you're a Python 2 person type:

    python server.py
If you're a Python 3 person type:

    python3 server.py
Some stuff will appear, but as long as it isn't very long it means you should be good. Head on over to a browser and in the URL field type:

localhost:3000
You should see something similar to that shown in Figure 50 below.

# Running the GUI
To run the GUI do the following:

    Make sure your Arduino is plugged into your computer, has the appropriate .ino sketch installed on it, and that it is not currently being used by any other piece of software. (For example, if you are using the Arduino IDE's built-in serial monitor, the Arduino will be "claimed" when you do the next step, preventing the 6.302 server software from being able to work with the Arduino.

    Start up your 6.302 server by typing either: python server.py or some variation into an appropriate terminal/shell (in Mac/UNix/Linux a standard terminal will work, in Windows, probalby it is best to use the PowerShell.)

    Open up a browser and in the url bar go to: localhost:3000. The GUI should then appear.

    In the top right, should be a pull-down tab that currently lists everything that "looks" like a valid serial port to the software. Find the correct port that corresponds to your Arduino.

    Below that is a BAUD selector that defaults to 115200. Unless we say otherwise, leave this as it is.

    Below that should be a "Connect" button. the "TX" LED on your Arduino shoudl start to flash and the plot in the lower right side should start to move. You should also note that the "Connect" Button now says "Connected (Click to Disconnect)".

    Parameters used by the Arduino can be changed "on-the-fly"...this was the whole purpose of the software actually. The sliders corresponding to the different parameters can either be moved manually or you can type in a value for more precision.

    The "Arduino Loop Headroom" piece of code can mostly be ignored...we just use it for debuggin purposes mainly. Hopefully this number should stay above 250 microseconds.

    The plot has three simple scaling buttons to zoom in/out/reset the y axis.

    The units on the x axis are a little hard to gauge (hence why we don't explicitely list them out). The box should be approximately about 5 seconds long, though this can vary because WebSockets aren't super consistent. That being said, in this class, this variation is sufficiently low for our purposes.

    When you want to Disconnect, press the button that now says "Disconnect". That should stop everything. If you want to restart, press the button again.

    If you want to turn off everything, go back into the terminal and press "Control+C" to kill the running Python Code.

Debugging
We've found this code to be pretty robust, but there will always be problems. One issue that has appeared is that the software will occasionally not be able to find your Arduino, or will occassionally, start communicating with the Arduino but not effectively transmit data onto the plot. In both cases, Disconnect, turn off your server (Control +C), unplug the Arduino, then try again. Reboots like this usually fix the problems.
